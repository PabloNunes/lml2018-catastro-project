```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
squared_meter_csv = pd.read_csv("valor.csv")
```

```python
squared_meter_csv
```

```python
localidad = squared_meter_csv["Localidad"].unique()
localidad
```

```python
schools_csv = pd.read_csv("Colegio.csv")
```

```python
schools_csv
```

```python
final_list = []
for index_school, row_school in squared_meter_csv.iterrows():
    x_school = row_school["coord_x"]
    y_school = row_school["coord_y"]
    for index_meter, row_meter in schools_csv.iterrows():
        house_list = []
        x_meter = row_meter["coord_x"]
        y_meter = row_meter["coord_y"]
        house_list.append(np.hypot(( float(x_school.replace(",", ".")) - float(x_meter.replace(",", ".")) ) , 
                             ( float(y_school.replace(",", ".")) - float(y_meter.replace(",", ".")) ) )  )
    final_list.append(min(house_list))
final_list
```

```python
plt.scatter(final_list, squared_meter_csv.iloc[:, 1:2])
plt.show()
```
